```python
Introduction
1.什么是SQL？
SQL全称是结构化查询语言，可以进行查询、操作、转换。SQL数据库包括SQLite、MySQL、Postgres、Oracle、Microsoft SQL Server，所有这些都支持通用的SQL语言标准，但附加功能和存储类型方面有所不同。其可以为上百万个网站提供安全且可扩展性存储。
2.什么是关系数据库？
关系数据库表示相关（二维）表的集合。每个表类似于Excel电子表格，具有固定数量的列和任意数量的数据行。
```

```python
Lesson4（distinct重复项、order排序、limit offset选择部分数据）
一、内容
1.Distinct：丢弃重复列
SELECT DISTINCT column,another column,... FROM mytable WHERE condition
由于DISTINCT关键字将盲目删除重复行，因此以后多用GROUP BY子句根据特定列丢弃重复项。
2.ORDER BY：按升序或降序对给定列排序结果
SELECT DISTINCT column,another column,... FROM mytable ORDER BY column ASC/DESC;
指定ORDER BY子句时，每行按照指定列的值进行字母数字排序。
在某些数据库中，您还可以指定排序规则以更好地对包含国际文本的数据进行排序。
3.LIMIT，OFFSET 
通常与ORDER BY子句一起使用的另一个子句是LIMIT和OFFSET子句，它们是一种有用的优化，用于向数据库指示您关注的结果的子集。
LIMIT：要返回的行数
OFFSET：将指定从哪里开始计算行数。
SELECT DISTINCT column,another column,... FROM mytable ORDER BY column ASC/DESC LIMIT num_limit OFFSET num_offset;
二、任务
1.List all directors of Pixar movies (alphabetically), without duplicates
2.List the last four Pixar movies released (ordered from most recent to least)
3.List the first five Pixar movies sorted alphabetically
4.List the next five Pixar movies sorted alphabetically
SELECT DISTINCT Director FROM movies ORDER BY Director ASC;
SELECT Title FROM movies ORDER BY Year DESC LIMIT 4 OFFSET 10 ;
SELECT Title FROM movies ORDER BY Title ASC LIMIT 5 OFFSET 0 ;
SELECT Title FROM movies ORDER BY Title ASC LIMIT 5 OFFSET 5;
```

```python
Lesson 6（inner join多表查询）
一、内容
1.数据规范化
数据库规范化很有用，因为它可以最大限度地减少任何单个表中的重复数据，并允许数据库中的数据彼此独立地增长。
2.共享有关单个实体的信息的表需要具有主键，该主键在数据库中唯一地标识该实体。 一种常见的主键类型是自动递增整数（因为它们节省空间），但它也可以是字符串，散列值，只要它是唯一的。
3.inner join=join：进行多表查询
SELECT DISTINCT column,another column,... FROM mytable INNER JOIN another_table ON mytable.id=another_table.id ORDER BY column ASC/DESC LIMIT num_limit OFFSET num_offset;
INNER JOIN是一个进程，它匹配第一个表和第二个表中具有相同键（由ON约束定义）的行，以创建包含两个表的组合列的结果行。
二、任务
1.Find the domestic and international sales for each movie
2.Show the sales numbers for each movie that did better internationally rather than domestically
3.List all the movies by their ratings in descending order
SELECT Title,Domestic_sales,International_sales FROM Boxoffice INNER JOIN Movies ON Movie_id = Id;
SELECT Title,Domestic_sales,International_sales FROM Boxoffice INNER JOIN Movies ON Movie_id = Id WHERE International_sales > Domestic_sales;
SELECT Title FROM Boxoffice INNER JOIN Movies ON Movie_id = Id ORDER BY Rating DESC;
```

```python
Lesson 10（聚合max count、聚合函数group by）
一、内容
1.除了我们上一课介绍的简单表达式之外，SQL还支持使用聚合表达式（或函数），这些表达式允许您汇总有关一组数据行的信息。 
SELECT AGG_FUNC(column_or_expression) AS aggregate_description, …
FROM mytable
WHERE constraint_expression;
如果没有指定的分组，每个聚合函数将在整个结果行集上运行并返回单个值。 与普通表达式一样，为聚合函数提供别名可确保结果更易于阅读和处理。
2.function
更多function：https://dev.mysql.com/doc/refman/5.6/en/group-by-functions.html
3.group by
除了聚合所有行之外，您还可以将聚合函数对指定字段进行分组。
SELECT AGG_FUNC(column_or_expression) AS aggregate_description, …
FROM mytable
WHERE constraint_expression
GROUP BY column;
4.注意：max、avg、sum可以分组之后计算每一组数据的最大、平均、求和，count求出一共有多少条数据。
二、任务
1.Find the longest time that an employee has been at the studio
2.For each role, find the average number of years employed by employees in that role
3.Find the total number of employee years worked in each building
1.SELECT max(Years_employed)
FROM employees;
2.SELECT role,avg(years_employed)
FROM employees
group by role
3.SELECT building,sum(years_employed)
FROM employees
group by building
```

```python
Lesson11（having：代替where的条件限定，专用于group）
一、内容
1.如果在WHERE子句之后执行GROUP BY子句，应该如何分组？
HAVING语句的存在弥补了WHERE关键字不能与聚合函数联合使用的不足。
注意：where必须在group之前。
2.HAVING子句约束的编写方式与WHERE子句约束相同，并应用于分组行。 如果你没有使用`GROUP BY`子句，一个简单的`WHERE`子句就足够了。
SELECT group_by_column, AGG_FUNC(column_expression) AS aggregate_result_alias, …
FROM mytable
WHERE condition
GROUP BY column
HAVING group_condition;
二、任务
1.Find the number of Artists in the studio (without a HAVING clause)
2.Find the number of Employees of each role in the studio
3.Find the total number of years employed by all Engineers
1.SELECT count(role)
FROM employees
where role="Artist"
2.SELECT role,count(role)
FROM employees
group by role
3.SELECT sum(years_employed) 
FROM employees
group by role
   having role="Engineer";
```

```python
Lesson 12（复习）
一、内容
1.组合
SELECT DISTINCT column, AGG_FUNC(column_or_expression), …
FROM mytable
    JOIN another_table
      ON mytable.column = another_table.column
    WHERE constraint_expression
    GROUP BY column
    HAVING constraint_expression
    ORDER BY column ASC/DESC
    LIMIT count OFFSET COUNT
2.顺序
查询的每个部分都是按顺序执行的，所以了解执行顺序非常重要。
   1）FROM和JOINs
首先执行FROM和JOIN，以确定正在查询的数据的总工作集。可以创建临时表，其中包含要连接的表的所有列和行。
2）WHERE
一旦我们拥有了总工作数据集，首先将WHERE约束应用于各个行，并且丢弃不满足约束的行。每个约束只能直接从FROM子句中请求的表访问列。无法从此步骤访问别名，因为还未执行select。
    3）GROUP BY
在WHERE约束之后的其余行将根据GROUP BY子句中指定的列中的公共值进行分组。只有值一样的才作为一行，这意味着只有在查询中具有聚合函数时才需要使用它。
4）HAVING
如果查询具有GROUP BY子句，则HAVING子句中的约束将应用在group之后，丢弃不满足约束的分组行。与WHERE子句一样，在大多数数据库中也无法从此步骤访问别名。
5）SELECT
最后计算查询的SELECT部分中的任何表达式。
    6）DISTINCT
在剩余的行中，标记为DISTINCT的列中具有重复值的行将被丢弃。
7）ORDER BY
如果ORDER BY子句指定了一个字段，那么这些行将按指定的数据按升序或降序排序。由于已经计算了查询的SELECT部分中的所有表达式，因此可以在此子句中引用别名。
8）LIMIT / OFFSET
最后，丢弃超出LIMIT和OFFSET指定范围的行，从查询返回最后一组行。
二、任务
1.Find the number of movies each director has directed
2.Find the total domestic and international sales that can be attributed to each director
1.SELECT director,count(title) 
FROM movies
group by director
2.SELECT director,sum(Domestic_sales+International_sales)
FROM movies
join boxoffice
where movies.id=boxoffice.movie_id
group by director
```

```python
Lesson 13（insert 增加）
一、内容
1.模式
我们之前将数据库中的表描述为一组二维的行和列，其中列是属性，行是表中实体的实例。 在SQL中，数据库模式描述了每个表的结构，以及表的每列可以包含的数据类型。
这种固定的结构允许数据库高效，并且尽管存储了数百万甚至数十亿行，但仍然保持一致。
2.Insert
将数据插入数据库时，我们需要使用INSERT语句，该语句声明要在哪个表进行insert。每行数据都应包含表中每个相应列的值。可以按顺序列出多个行，一次插入多行。
INSERT INTO mytable
VALUES (value_or_expr, another_value_or_expr, …),
       (value_or_expr_2, another_value_or_expr_2, …),
       …;
3.insert（指定列）
在某些情况下，如果您的数据不完整且表中包含默认值的列，则可以通过显式指定行来插入仅包含数据列的行。在这些情况下，值的数量需要与指定的列数相匹配。 
尽管这是一个更冗长的写法陈述，但以这种方式插入值具有兼容的好处。例如，如果使用默认值向表中添加新列，则不需要因此更改编码的INSERT语句。
INSERT INTO mytable
(column, another_column, …)
VALUES (value_or_expr, another_value_or_expr, …),
      (value_or_expr_2, another_value_or_expr_2, …),
      …;
4.Insert（字符串数字一起插入）
此外，您可以将数学和字符串表达式一起插入。以确保插入的所有数据都以某种方式格式化。
INSERT INTO boxoffice
(movie_id, rating, sales_in_millions)
VALUES (1, 9.9, 283742034 / 1000000);
二、任务
1.Add the studio's new production, Toy Story 4 to the list of movies (you can use any director)
2.Toy Story 4 has been released to critical acclaim! It had a rating of 8.7, and made 340 million domestically and 270 million internationally. Add the record to the BoxOffice table.
1.Insert into movies
(title)
Values(Toy Story 4)
2.Insert into boxoffice
Values(15,8.7,340000000,270000000)
```

```python
Lesson 14（update）
一、内容
1.update
除了添加新数据之外，常见的任务是更新现有数据，这可以使用UPDATE语句来完成。 与INSERT语句类似，您必须准确指定要更新的表（update），行（where）和对应行的列的数值（set）。 此外，要更新的数据必须与表模式中列的数据类型相匹配。
UPDATE mytable
SET column = value_or_expr, 
    other_column = another_value_or_expr, 
    …
WHERE condition;
2.注意
大多数使用SQL的人都会犯错误。所以，一个有用的提示是始终首先编写约束（where）并在SELECT查询中对其进行测试，以确保更新正确的行，然后再写入列/值对进行更新。
二、任务
1.The director for A Bug's Life is incorrect, it was actually directed by John Lasseter
2.The year that Toy Story 2 was released is incorrect, it was actually released in 1999
3.Both the title and directory for Toy Story 8 is incorrect! The title should be "Toy Story 3" and it was directed by Lee Unkrich
1.
Update movies
Set director="John Lasseter"
Where title ="A Bug's Life"
2.
Update movies
Set year=1999
Where title ="Toy Story 2"
3.
Update movies
Set title="Toy Story 3",
    director="Lee Unkrich"
Where id=11
```
